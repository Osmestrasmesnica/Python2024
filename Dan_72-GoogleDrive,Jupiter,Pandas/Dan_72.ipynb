# INTRO

```python
import pandas as pd
df = pd.read_csv('salaries_by_college_major.csv')
df.head()
```

To see the number of rows and columns we can use the shape attribute

```python
df.shape
```

We can access the column names directly with the columns attribute.

```python
df.columns
```

Use the .isna() method and see if you can spot if there's a problem somewhere.


```python
df.isna()
```

Check the last couple of rows

```python
df.tail()
```

Let's create a new dataframe without the last row and examine the last 5 rows to make sure we removed the last row:

```python
clean_df = df.dropna()
clean_df.tail()
```

# Find College Major with Highest Starting Salaries

```python
clean_df['Starting Median Salary']
```

To find the highest starting salary we can simply chain the .max() method.

```python
clean_df['Starting Median Salary'].max()
```

The .idxmax() method will give us index for the row with the largest value.

```python
clean_df['Starting Median Salary'].idxmax()
```

To see the name of the major that corresponds to that particular row, we can use the .loc (location) property

```python
clean_df['Undergraduate Major'].loc[43]
```

using the double square brackets notation to achieve exactly the same thing:

```python
clean_df['Undergraduate Major'][43]
```

```python
clean_df.loc[43]
```

# What college major has the highest mid-career salary? How much do graduates with this major earn? (Mid-career is defined as having 10+ years of experience).

```python
print(clean_df["Mid-Career Median Salary"].idxmax())
clean_df.loc[8]
```

# Which college major has the lowest starting salary and how much do graduates earn after university?

```python
clean_df["Starting Median Salary"].idxmin()
clean_df.loc[49]
```

# Which college major has the lowest mid-career salary and how much can people expect to earn with this degree?

```python
clean_df["Mid-Career Median Salary"].idxmin()
clean_df.loc[18]
```

# How would we calculate the difference between the earnings of the 10th and 90th percentile? Pandas allows us to do simple arithmetic with entire columns, so all we need to do is take the difference between the two columns:

```python
# Version 1
clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
```

```python
# Version 2
clean_df['Mid-Career 90th Percentile Salary'].subtract(clean_df['Mid-Career 10th Percentile Salary'])
```

To add new column we use .insert() method

```python
spread_col = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1, 'Spread', spread_col) # Number 1 - where the column should be inserted (at position 1, so the second column, it starts from 0)
clean_df.head()
```

Sorting by the Lowest Spread .sort_values()

```python
low_risk = clean_df.sort_values('Spread') # sort_values() is by default ascending = True
low_risk[['Undergraduate Major', 'Spread']].head()
```

# Find the top 5 degrees with the highest values in the 90th percentile.

```python
highest_potential  = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
highest_potential[["Undergraduate Major", "Mid-Career 90th Percentile Salary"]].head()
```

# Find the degrees with the greatest spread in salaries

```python
highest_spread  = clean_df.sort_values('Spread', ascending = False) # sort_values() is by default ascending = True
highest_spread[['Undergraduate Major', 'Spread']].head()
```

```python
highest_median = clean_df.sort_values('Mid-Career Median Salary', ascending = False)
highest_median[['Undergraduate Major', 'Mid-Career Median Salary']].head(10)
```

# Which category of degrees has the highest average salary? Is it STEM, Business or HASS (Humanities, Arts, and Social Science)? We use .groupby() and count()

```python
clean_df.groupby('Group').count()
```

# Find the average salary by group?



```python
clean_df.groupby('Group').mean()
```
